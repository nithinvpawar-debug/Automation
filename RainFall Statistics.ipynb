```python
mean_annual_rainfall = df.groupby('YEAR')['ANNUAL'].mean()
mean_annual_rainfall.plot(legend=True, label='mean annual rainfall', figsize=(12, 10))
plt.xlabel('Year')
plt.ylabel('Mean Annual Rainfall')
plt.title('Mean Annual Rainfall Over the Years')
plt.show()
```

```python
mean_annual_rainfall = df.groupby('YEAR')['ANNUAL'].mean()
mean_annual_rainfall.plot(legend=True, label='Mean Annual Rainfall', figsize=(12, 10))
ma10 = mean_annual_rainfall.rolling(10).mean()
ma10.plot(legend=True, label='Moving Avg. for last 10 years')
plt.xlabel('Year', fontsize=20)
plt.ylabel('Annual Rainfall (in mm)', fontsize=20)
plt.title('Annual Rainfall in India from year 1901-2015', fontsize=25)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(23, 10))
ax = sns.boxplot(y='SUBDIVISION', x='ANNUAL', data=df, width=0.8)
ax.set_xlabel('Annual Rainfall in mm', fontsize=30)
plt.show()
```

# Observations

Above two graphs shows that the amount of rainfall is resonably good in the months of march, april, may in estern India 

```python
plt.figure(figsize=(10,5))
sns.heatmap(df[['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec', 'ANNUAL']].corr(),annot=True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL','AUG', 'SEP', 'OCT', 'NOV', 'DEC','ANNUAL']].corr(),annot=True)
plt.show()
```

# Observations

* Heat map shows the co-relation(depedency)between the amounts of rainfall over months.
* From above it is clear that if amount of rainfall is high in the months. 

```python
andaman_nicobar_islands = df.loc[df.SUBDIVISION == 'SOUTH INTERIOR KARNATAKA', :]
andaman_nicobar_islands.groupby("YEAR").ANNUAL.mean().plot(legend=True, label='andaman_nicobar_islands')
mean_annual_rainfall = df.groupby('YEAR').ANNUAL.mean()
mean_annual_rainfall.plot(figsize=(12, 8), title='Annual rainfall in andaman_nicobar_islands compared to the whole of India', legend=True, label='India')
plt.ylabel('Annual rainfall in mm')
plt.show()
```
